```python
import pandas as pd
from sklearn.linear_model import LinearRegression
URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-BD0231EN-SkillsNetwork/datasets/diamonds.csv"
df=pd.read_csv(URL)
```

```python
df.head()
```

```python
df.plot.scatter(x='carat',y='price')
```

```python
df.plot.scatter(x='depth',y='price')
```

 As you can see the depth feature isnt forming straight line hence its weak feature for price target.

```python
df.plot.scatter(x='x',y='price')   #has few outliner!
```

```python
df=df[df['x'] > 0]
```

```python
df.plot.scatter(x='x',y='price')
```

```python
target=df["price"]
features=df[["carat","x"]]
```

```python
lr=LinearRegression()
lr.fit(features,target)
```

```python
lr.score(features,target)
```
